# Top 15 countries under threat

```python
df['country_txt'].value_counts()
```

```python
plt.figure(figsize=(20,8))
c = df['country_txt'].value_counts().index
sns.countplot(df['country_txt'], order=c[:16]);
plt.xlabel('Countries')
plt.ylabel('Death rate')
plt.title('Top 15 Countries under threat')
plt.show()
```

# Types of attacks

```python
df['attacktype1_txt'].value_counts()
```

```python
plt.figure(figsize=(20,8))
attack = df['attacktype1_txt'].value_counts().index
sns.countplot(df['attacktype1_txt'], order=attack[:]);
plt.xlabel('Attack type')
plt.ylabel('Attack count')
plt.title('Types of attacks')
plt.show()
```

# Terrorism activity per year

```python
plt.subplots(figsize=(15,6))
sns.countplot('iyear', data=df)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.title('Terrorism Activities per year')
plt.show()
```

# Top 10 most affected cities

```python
df['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Cities",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top 10 most affected city",fontsize=20)
plt.show()
#Few group names are not revealed, thus we call them unknown.
```
